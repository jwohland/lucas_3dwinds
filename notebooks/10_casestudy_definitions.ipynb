```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
ds_oro = xr.open_dataset("../data/JLU/oro/oro_EUR44_ECMWF-ERAINT_LUCAS_FOREST_JLU_CCLM-5-0-9_fx.nc")
```

```python
ds_oro["HSURF"].where(ds_oro["HSURF"]>1).plot(cmap="RdBu", levels=[0,100,200,400,800,1600,3200])
```

```python
def plot_section(rlat_0, rlat_1, rlon_0, rlon_1, ds_oro=ds_oro, title="", name="test"):
    rlats = slice(rlat_0, rlat_1)
    rlons = slice(rlon_0, rlon_1)
    f, ax = plt.subplots()
    ds_oro["HSURF"].where(ds_oro["HSURF"]>1).sel(rlat=rlats, rlon=rlons).plot(cmap="RdBu",levels=[0,100,200,400,800,1600,3200])
    ax.set_title(title)
    plt.savefig("../plots/focus_regions/" + name + ".jpeg", dpi=300)
```

```python
plot_section(-20,22,-25,18, name="Europe")
```

```python
plot_section(8, 23, -20, -5, title="North of UK", name="NorthSea")
```

```python
plot_section(-12, -8, -15, -9, title="Baleares", name="Baleares")
```

```python
plot_section(-20, -10, 0, 10, title="Crete", name = "Crete")
```

```python
plot_section(-20, -10, 10, 15, title="Cyprus")
```
