The horizontal coordinates in BCCR data is not always the same:

va EVAL 1986 1000 hPa data has x and y coordinates instead of rlat & rlat #27 

```python
import xarray as xr
from os import listdir
```

```python
data_path = "../data/BCCR/"
experiment = "EVAL"
```

```python
filenames = listdir(data_path + experiment + "/raw_data")
```

```python
ds_list = [xr.open_dataset(data_path + experiment + "/raw_data/" + x) for x in filenames if "1986010100-" in x]
```

```python
for i in range(len(ds_list)):
    print(ds_list[i].dims)
```

```python
# Repeat scanning for all years and all experiments
problem_list = []
for experiment in ["EVAL", "FOREST", "GRASS"]:
    filenames = listdir(data_path + experiment + "/raw_data")
    for year in range(1986,2016):
        ds_list = [xr.open_dataset(data_path + experiment + "/raw_data/" + x) for x in filenames if str(year) + "010100-" in x]
        for ds in ds_list:
            if "rlon" not in ds.dims:
                print("Issue in " + experiment + " in year " + str(year))
                problem_list.append(ds)
```
